```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\AILab\canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per capita income (US$')
plt.scatter(df['year'],df['per capita income (US$)'],color = 'red',marker = '+')
```

```python
new_df  =df.drop('per capita income (US$)',axis = 'columns')
new_df
```

```python
capita = df['per capita income (US$)']
capita
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,capita)
```

# Pedict the per capita income of canadian citizen in year 2020

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
